```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
cust_df=pd.DataFrame(df1)
cust_df
```

```python
trans_df=pd.DataFrame(df2)
trans_df
```

```python
cust_df.columns
trans_df.columns
```

```python
#transforming the customers dataframe
cust_df=cust_df[['customer_id','customer_name','country','city']]
#transforming products dataframe
products_df=pd.DataFrame({
    'product_id':trans_df['product_id'].unique(),
    'product_name':trans_df.drop_duplicates('product_id')['product_name'],
    'product_category':trans_df.drop_duplicates('product_id')['product_category']
})
products_df
```

```python
df2.columns
```

```python
ecommerceweb_df=pd.DataFrame({
    'website_id':range(len(trans_df['ecommerce_website_name'].unique())),
    'website_name':trans_df['ecommerce_website_name'].unique()
})
ecommerceweb_df
```

```python
range(len(trans_df['ecommerce_website_name'].unique()))
```

```python
websitename_toid=ecommerceweb_df.set_index('website_name')['website_id'].to_dict()
websitename_toid
```

```python
trans_df.columns
```

```python
trans_df['website_id']=trans_df['ecommerce_website_name'].map(websitename_toid)
```

```python
trans_df['datetime']=pd.to_datetime(trans_df['datetime'])
trans_df=trans_df[['order_id','customer_id','product_id','website_id','qty','price','datetime','payment_type','payment_txn_id','payment_txn_success','failure_reason']]
```

```python
!pip install pymysql
```

```python
#mysql+pymysql specifies the datbase ('mysql' and the python dbapi ('pymysql')
#last one is the database name
engine=create_engine('mysql+pymysql://root:root@localhost/rev_p0')
```

```python
cust_df
products_df
ecommerceweb_df
trans_df
```

```python
cust_df.to_sql('customers',con=engine,if_exists='replace',index=False)
```

```python
products_df.to_sql('products',con=engine,if_exists='replace',index=False)
```

```python
ecommerceweb_df.to_sql('ecommerce_websites',con=engine,if_exists='replace',index=False)
```

```python
trans_df.to_sql('transactions',con=engine,if_exists='replace',index=False)
```
